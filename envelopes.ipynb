Doel: Een zo groot mogelijke stapel enveloppen in elkaar steken. De hoogte en
breedte van elke envelop is bekend. Een envelop past in een andere envelop als
de hoogte en de breedte kleiner is. Enveloppen mogen niet worden gedraaid.
Op Brightspace staat een bestand ‘enveloppen.txt’ met hoogtes en breedtes van 50
enveloppen.
Verzin een algoritme dat voor een willekeurige verzameling enveloppen de grootst
mogelijke stapel bepaalt.
Verzin een efficiënt algoritme dat voor een willekeurige verzameling enveloppen de
grootst mogelijke stapel bepaalt

```python
#Using Longest Increasing Subsequence (LIS) after sorting the envelopes, we can find the maximum number of envelopes that can be nested inside each other.
#time complexity: O(n log n) due to sorting and the LIS algorithm, where n is the number of envelopes.
def max_envelopes(envelopes):
    # Sort the envelopes by width in ascending order and by height in descending order for equal widths
    envelopes.sort(key=lambda x: (x[0], -x[1]))

    # Extract the heights of the sorted envelopes
    heights = [h for _, h in envelopes]

    # Function to find the length of the longest increasing subsequence
    def length_of_lis(nums):
        from bisect import bisect_left #bisect_left is a function from the bisect module that finds the position where an element should be inserted to maintain sorted order.
        lis = []
        for num in nums:
            pos = bisect_left(lis, num)
            if pos == len(lis):
                lis.append(num)
            else:
                lis[pos] = num
        return len(lis)

    # Find the length of the longest increasing subsequence of heights
    return length_of_lis(heights)
print(max_envelopes(envelopes))

#if we want to return the actual sequence of envelopes instead of just the length, we can modify the algorithm to keep track of the indices of the envelopes in the longest increasing subsequence.
def max_envelopes_with_sequence(envelopes):
    # Sort the envelopes by width in ascending order and by height in descending order for equal widths
    envelopes.sort(key=lambda x: (x[0], -x[1]))

    # Extract the heights of the sorted envelopes
    heights = [h for _, h in envelopes]

    # Function to find the longest increasing subsequence and return the sequence of indices
    def length_of_lis(nums):
        from bisect import bisect_left
        lis = []
        parent = [-1] * len(nums)
        for i, num in enumerate(nums): #enumerate is used to get both the index and the value of each element in the heights list.
            pos = bisect_left(lis, num)
            if pos == len(lis):
                lis.append(num)
            else:
                lis[pos] = num
            if pos > 0:
                parent[i] = lis[pos - 1]
        # Reconstruct the longest increasing subsequence
        sequence = []
        k = lis[-1]
        for i in range(len(nums) - 1, -1, -1):
            if nums[i] == k:
                sequence.append(i)
                k = parent[i]
        return sequence[::-1]

    # Get the indices of the longest increasing subsequence of heights
    sequence_indices = length_of_lis(heights)

    # Return the actual sequence of envelopes based on the indices
    return [envelopes[i] for i in sequence_indices]
print(max_envelopes_with_sequence(envelopes))
```
