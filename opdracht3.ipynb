1 Schrijf een (recursieve) methode ’subList(self, start, end)’, die een nieuwe lijst retourneert met alle elementen die
in de oorspronkelijke lijst tussen positie ’start’ (inclusief) en ’end’ (exclusief) liggen. Een aanroep bijvoorbeeld van
’subList(1, 3)’ op een lijst met elementen ’5’, ’4’, ’7’ en ’4’ levert een nieuwe lijst op met elementen ’4’ en ’7’. Schrijf
tests die de werking van de methode aannemelijk maken.

```python
#sublist recursive function
'''
#non-recursive version
def subList(self, start, end):
    return self[start:end]
'''

def subList(self, start, end):
    if start >= end:
        return []
    else:
        return [self[start]] + subList(self, start + 1, end)



#test cases
my_list = [1, 2, 3, 4, 5]
print(subList(my_list, 0, 2)) #should print [1, 2]
print(subList(my_list, 2, 4)) #should print [3, 4]
print(subList(my_list, 1, 3)) #should print [2, 3]
print(subList(my_list, 0, 4)) #should print [1, 2, 3, 4]
print(subList(my_list, 3, 3)) #should print [] (empty list)
```

3 Schrijf een (recursieve) methode ’merge(self, list)’, die een nieuwe lijst retourneert met alle elementen uit de oor-
spronkelijke en de gegeven lijst samengevoegd en gesorteerd op volgorde van klein naar groot. Er mag vanuit gegaan
worden dat de oorspronkelijke en de gegeven lijst beide gesorteerd zijn van klein naar groot. Een aanroep bijvoor-
beeld van ’merge(list)’ op een lijst met elementen ’4’, ’4’, ’5’ en ’7’ en waarbij ’list’ een lijst is met elementen ’2’, ’6’
en ’7’ levert een nieuwe lijst op met elementen ’2’, ’4’, ’4’, ’5’, 6’, ’7’ en ’7’. Bepaal zelf in hoeverre je van recursie
en/of hulpmethoden gebruik wilt maken. Schrijf tests die de werking van de methode aannemelijk maken.

```python
#recursive function merge(self, list)
'''
#time complexity: O(n log n) where n is the total number of elements in both lists (self + list)
def merge(self, list):
    if not list:
        return self
    #merged and sorted low to high
    else:
        merged = sorted(self + list)
        return merged
'''

def merge(self, list):
    merged = []
    i, j = 0, 0
    while i < len(self) and j < len(list):
        if self[i] < list[j]:
            merged.append(self[i])
            i += 1
        else:
            merged.append(list[j])
            j += 1
    # Add remaining elements from self or list
    merged.extend(self[i:])
    merged.extend(list[j:])
    return merged


#test cases
list1 = [1, 3, 5]
list2 = [2, 4]
print(merge(list1, list2)) #should print [1, 2, 3, 4, 5]
list3 = [7, 8]
print(merge(list1, list3)) #should print [1, 3, 5, 7, 8]
list4 = [0, 5, 6]
print(merge(list1, list4)) #should print [0, 1, 3, 5, 5, 6]
list5 = []
print(merge(list1, list5)) #should print [1, 3, 5] (same as list1)
```

4 Wat is de tijdscomplexiteit van jouw ’merge’ implementatie? Onderbouw dit.

```python
#there is 3 possible cases for merging two lists:
#1. Both lists are sorted: time complexity is O(n) where n is the total number of elements in both lists.

#2. One list is sorted and the other is not: time complexity is O(n) where n is the total number of elements in both lists.
#   - If the sorted list is the first list (self), we can iterate through it and merge it with the unsorted list in O(n) time.
#   - If the sorted list is the second list (list), we can iterate through it and merge it with the unsorted list in O(n) time.

#3. Neither list is sorted: time complexity would be O(n log n).
```
