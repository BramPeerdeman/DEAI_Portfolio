Geef met behulp van bovenstaande hulpcode een overzicht met daarin op basis van het aantal jaar dat iemand in dienst is of een medewerker ‘kort in dienst’ (minder dan 30 jaar in dienst) of een ‘lang in dienst’ (groter dan gelijk aan 30 jaar in dienst) is. Geef daarbij per medewerker in een aparte kolom zowel ‘kort in dienst’ als ‘lang in dienst’ aan. Gebruik (wederom) een for-loop.<br>
(2 kolommen, 102 rijen) 

```python

sales_staff["DATE_HIRED"] = pd.to_datetime(sales_staff["DATE_HIRED"], format='%Y-%m-%d', errors='coerce')
huidig_jaar = date.today().year
sales_staff["jaar_in_dienst"] = (huidig_jaar - sales_staff["DATE_HIRED"].dt.year).fillna(0).astype(int)
sales_staff["Dienstverband"] = sales_staff["jaar_in_dienst"].apply(lambda x: "Kort in dienst" if x < 30 else "Lang in dienst")
resultaat = sales_staff[["jaar_in_dienst", "Dienstverband"]].sort_values(by="jaar_in_dienst", ascending=False)

resultaat
```
