## Databasetabellen inlezen

Kopieer in het codeblok hieronder van het vorige werkcollege de code om een connectie met het bestand go_sales.sqlite te maken.

```python
import sqlite3
conn = sqlite3.connect("data/go_sales.sqlite")
print("Databaseconnectie succesvol!")
```

Lees van de ingelezen go_sales-database te volgende tabellen in met behulp van "SELECT * FROM *tabel*".
- product
- product_type
- product_line
- sales_staff
- sales_branch
- retailer_site
- country
- order_header
- order_details
- returned_item
- return_reason

```python
product = "SELECT * FROM product"
df_producten = pd.read_sql(product, conn)
product_type = "SELECT * FROM product_type"
df_product_type = pd.read_sql(product_type, conn)
product_line = "SELECT * FROM product_line"
df_product_line = pd.read_sql(product_line, conn)
sales_staff = "SELECT * FROM sales_staff"
df_sales_staff = pd.read_sql(sales_staff, conn)
sales_branch = "SELECT * FROM sales_branch"
df_sales_branch = pd.read_sql(sales_branch, conn)
retailer_site = "SELECT * FROM retailer_site"
df_retailer_site = pd.read_sql(retailer_site, conn)
country = "SELECT * FROM country"
df_country = pd.read_sql(country, conn)
order_header = "SELECT * FROM order_header"  
df_order_header = pd.read_sql(order_header, conn)
order_details = "SELECT * FROM order_details"
df_order_details = pd.read_sql(order_details, conn)
returned_item = "SELECT * FROM returned_item"
df_returned_item = pd.read_sql(returned_item, conn)
return_reason = "SELECT * FROM return_reason"
df_return_reason = pd.read_sql(return_reason, conn)

print("Tabellen succesvol ingeladen in Pandas DataFrames!")
```

Geef een overzicht met daarin de productnamen waarvan ooit meer dan 40% korting is verleend. De formule voor korting is: (unit_price - unit_sale_price) / unit_price (1 kolom, 8 rijen) 

```python
df_merged = pd.merge(df_producten, df_order_details, on='PRODUCT_NUMBER')

discount_con = ((df_merged['UNIT_PRICE'] - df_merged['UNIT_SALE_PRICE']) / df_merged['UNIT_PRICE']) > 0.40

resultaat = df_merged.loc[discount_con, ['PRODUCT_NAME']].drop_duplicates()

print(resultaat)
```

Geef een overzicht met daarin van de producten behorend tot het producttype ‘Cooking Gear’ per productnaam de totaal verkochte hoeveelheid en de gemiddelde verkoopprijs. Sorteer de uitkomst op totaal verkochte hoeveelheid. (6 kolommen, 10 rijen)

```python
df_p_pt = pd.merge(df_producten, df_product_type, on='PRODUCT_TYPE_CODE')
df_cooking = df_p_pt[df_p_pt['PRODUCT_TYPE_EN'] == 'Cooking Gear']

df_total = pd.merge(df_cooking, df_order_details, on='PRODUCT_NUMBER')

resultaat = df_total.groupby(
[
    'PRODUCT_NUMBER', 
    'PRODUCT_NAME', 
    'PRODUCT_TYPE_CODE', 
    'PRODUCT_TYPE_EN'
]).agg(
{
    'QUANTITY': 'sum',
    'UNIT_SALE_PRICE': 'mean'
}).reset_index()

resultaat.columns = [
    'PRODUCT_NUMBER', 'PRODUCT_NAME', 'PRODUCT_TYPE_CODE', 
    'PRODUCT_TYPE_EN', 'TOTAAL_VERKOCHT', 'GEM_VERKOOPPRIJS'
]

resultaat = resultaat.sort_values(by='TOTAAL_VERKOCHT', ascending=False)

print(resultaat)
```

Geef een overzicht met daarin per land de naam van het land, de naam van de stad waar de verkoopafdeling is gevestigd (noem de kolomnaam in het overzicht ‘verkoper’) en het aantal steden waar zich klanten bevinden in dat land (noem de kolomnaam in het overzicht ‘klanten’) (3 kolommen, 29 rijen) 

```python
klanten_per_land = df_retailer_site.groupby('COUNTRY_CODE')['CITY'].nunique().reset_index()
klanten_per_land.columns = ['COUNTRY_CODE', 'klanten']

df_verkoper = pd.merge(df_sales_branch, df_country, on='COUNTRY_CODE')

df_final = pd.merge(df_verkoper, klanten_per_land, on='COUNTRY_CODE')

resultaat = df_final[['COUNTRY', 'CITY', 'klanten']].copy()
resultaat.columns = ['COUNTRY', 'verkoper', 'klanten']

print(resultaat)
```
