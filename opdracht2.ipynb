1 Maak een eigen implementatie van een ’Singly linked list’. Er zijn meerdere implementaties denkbaar, maar we
adviseren om de list als abstracte class te implementeren met twee concrete subclasses. Een voor lege en een voor
niet-lege lijsten.

```python
#single linked list implementation with abstract class and subclasses for empty and non-empty nodes
#abstract class Node:
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None

#subclass for empty list
class EmptyNode(Node):
    def __init__(self):
        super().__init__(None)

#subclass for non-empty list
class NonEmptyNode(Node):
    def __init__(self, value):
        super().__init__(value)
```

2 Schrijf een (recursieve) methode ’toString’, die een lijst omzet naar een string waarin de inhoud wordt opgesomt
gescheiden door een spatie. (Een overtollige spatie aan het einde is niet erg.) Een lijst met elementen ’4’ en ’7’
bijvoorbeeld, zou omgezet worden naar ”4 7 ”. Schrijf voorbeeldcode met lijsten van 0, 1 en 2 elementen.

```python
#a recursive function toString to prin the list, where elements are concatenated with spaces
#be careful, NoneType has no attribute 'value' or 'next', so we need to check for that in the function
def toString(node):
    if node is None or isinstance(node, EmptyNode):
        return ""
    else:
        return str(node.value) + " " + toString(node.next)
#exapmle code for lists with 0, 1 and 2 elements
empty_list = EmptyNode()
print("Empty list: " + toString(empty_list))
single_element_list = NonEmptyNode(1)
print("Single element list: " + toString(single_element_list))
two_element_list = NonEmptyNode(1)
two_element_list.next = NonEmptyNode(2)
print("Two element list: " + toString(two_element_list))
```

3 Schrijf een (recursieve) methode ’addFirst(value), die een nieuwe lijst retourneert met de gegeven waarde vooraan
en daarna alle elementen uit de oorspronkelijke lijst. Een aanroep van ’addFirst(5)’ op een lijst met elementen ’4’ en
’7’ bijvoorbeeld levert een nieuwe lijst op met elementen ’5’, ’4’ en ’7’. Schrijf voorbeeldcode die aannemelijk maakt
dat de methode correct werkt.

```python
#a recurse function addFirst(value) that adds a new node with the given value at the beginning of the list and returns the new head of the list
def addFirst(node, value):
    new_node = NonEmptyNode(value)
    new_node.next = node
    return new_node
#example code for adding elements to the list
list_with_three_elements = addFirst(two_element_list, 0)
print("List with three elements: " + toString(list_with_three_elements))
```

4 Schrijf een (recursieve) methode ’remove(value), die een nieuwe lijst retourneert met alle elementen uit de oor-
spronkelijke lijst, behalve het eerste voorkomen van de gegeven waarde. Een aanroep van ’remove(4)’ op een lijst
met elementen ’5’, ’4’, ’7’ en ’4’ bijvoorbeeld levert een nieuwe lijst op met elementen ’5’, ’7’ en ’4’. Een aanroep
van ’remove(4)’ op deze laatste lijst levert een nieuwe lijst op met elementen ’5’ en ’7’. Schrijf voorbeeldcode die
aannemelijk maakt dat de methode correct werk

```python
#a recursive function remove(value) that returns a new list with the first occurrence of the given value removed. If the value is not found, the original list should be returned.
def remove(node, value):
    if node is None or isinstance(node, EmptyNode):
        return node
    elif node.value == value:
        return node.next
    else:
        node.next = remove(node.next, value)
        return node
#example code for removing elements from the list
list_with_two_elements = remove(list_with_three_elements, 1)
print("List with two elements: " + toString(list_with_two_elements))
```

Er is nog een kleine hulpmethode nodig. Schrijf een (recursieve) methode ’smallest()’ die de kleinste waarde van
een lijst retourneert. Een aanroep van ’smallest()’ op een lijst met elementen ’5’, ’4’ en ’7’ bijvoorbeeld geeft ’4’ als antwoord. Schrijf voorbeeldcode die aannemelijk maakt dat de methode correct werkt.

```python
#now, we will find the smallest element in the original list and remove it, Then we can add it to a new list and repeat this process until the original list is empty. This way we can sort the list in ascending order.
#first we need to find the smallest element in the list, we can do this by iterating through the list and keeping track of the smallest value we have seen so far.
def findSmallest(node):
    if node is None or isinstance(node, EmptyNode):
        return None
    smallest = node.value
    current = node.next
    while current is not None and not isinstance(current, EmptyNode):
        if current.value < smallest:
            smallest = current.value
        current = current.next
    return smallest
#example code for finding the smallest element in the list
smallest_element = findSmallest(list_with_three_elements)
```

6 Alle ingredienten zijn aanwezig voor het eenvoudige sorteeralgoritme! Schrijf een (recursieve) methode ’sortSimple(),
die een nieuwe lijst retourneert met alle elementen uit de oorspronkelijke lijst gesorteerd van klein naar groot. Een
aanroep van ’sortSimple()’ op een lijst met elementen ’5’, ’4’, ’7’ en ’4’ bijvoorbeeld levert een nieuwe lijst op met
elementen ’4’, ’4’, ’5’ en ’7’. Schrijf voorbeeldcode die aannemelijk maakt dat de methode correct werkt.
Mocht een recursieve methode te lastig zijn mag een niet-recursieve methode ook. Het kan dan handig zijn om
gebruik te maken van een hulpmethode ’largest()’ in plaats van ’smallest()’.

```python
#now we can use the findSmallest function to sort the list, we can do this by creating a new list and adding the smallest element to it, then we can remove the smallest element from the original list and repeat this process until the original list is empty.
def sortSimple(node):
    if node is None or isinstance(node, EmptyNode):
        return node
    smallest = findSmallest(node)
    new_list = NonEmptyNode(smallest)
    new_list.next = sortSimple(remove(node, smallest))
    return new_list

#example code for sorting the list
unsorted_list = NonEmptyNode(3)
unsorted_list.next = NonEmptyNode(1)
unsorted_list.next.next = NonEmptyNode(4)
unsorted_list.next.next.next = NonEmptyNode(2)
print("Unsorted list: " + toString(unsorted_list))
sorted_list = sortSimple(unsorted_list)
print("Sorted list: " + toString(sorted_list))
```

8 Schrijf een (recursieve) methode ’uniq()’die, indien aangeroepen op een gesorteerde lijst, retourneert hoeveel unieke
elementen de lijst bevat. Een aanroep van ’uniq()’ op een lijst met elementen ’4’, ’4’, ’5’ en ’7’ bijvoorbeeld geeft ’3’
als antwoord. Schrijf voorbeeldcode die aannemelijk maakt dat de methode correct werkt.

```python
#a recursive function uniq() that, when called on a sorted list, resturns how manu unique elements are in the list. For example, if the list contains the elements 1, 1, 2, 3, 3, 3, 4, then uniq() should return 4, because there are four unique elements in the list: 1, 2, 3 and 4.
def uniq(node):
    if node is None or isinstance(node, EmptyNode):
        return 0
    count = 1
    current = node.next
    while current is not None and not isinstance(current, EmptyNode):
        if current.value != node.value:
            count += 1
            node = current
        current = current.next
    return count

#example code for counting the unique elements in the list
sorted_list_with_duplicates = NonEmptyNode(1)
sorted_list_with_duplicates.next = NonEmptyNode(1)
sorted_list_with_duplicates.next.next = NonEmptyNode(2)
sorted_list_with_duplicates.next.next.next = NonEmptyNode(3)
sorted_list_with_duplicates.next.next.next.next = NonEmptyNode(3)
sorted_list_with_duplicates.next.next.next.next.next = NonEmptyNode(3)
sorted_list_with_duplicates.next.next.next.next.next.next = NonEmptyNode(4)
unique_count = uniq(sorted_list_with_duplicates)
print("Number of unique elements: " + str(unique_count))
```
