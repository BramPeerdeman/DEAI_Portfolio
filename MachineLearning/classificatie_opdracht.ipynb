```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix

df = pd.read_excel('AmesHousing.xlsx') 

# Encoden van de categorische feature (House Style)
df_encoded = pd.get_dummies(df, columns=['House Style'], drop_first=True)

# Target bepalen: Dit is een ja/nee vraag (Binair)
y1 = df_encoded['Garage']
feature_cols1 = ['Year Built', 'Gr Liv Area'] + [col for col in df_encoded.columns if 'House Style_' in col]
X1 = df_encoded[feature_cols1]

# Splitsen van data in 80% train en 20% test
X_train1, X_test1, y_train1, y_test1 = train_test_split(X1, y1, test_size=0.2, random_state=42)

# Initialiseren van de Decision Tree. 
# max_depth beperkt hoeveel vragen de boom mag stellen om overfitting te voorkomen.
dt_model1 = DecisionTreeClassifier(max_depth=5, min_samples_split=10, random_state=42)
dt_model1.fit(X_train1, y_train1)

# Evalueren met classificatie-metrieken: Accuracy en de Confusion Matrix
y_pred1 = dt_model1.predict(X_test1)
accuracy1 = accuracy_score(y_test1, y_pred1)
conf_matrix1 = confusion_matrix(y_test1, y_pred1)

print(f"Accuracy: {accuracy1 * 100:.2f}%")
print("Confusion Matrix:\n", conf_matrix1)
```

```python
# Meer features toevoegen om huizen zonder garage beter te herkennen
feature_cols_exp1 = ['Year Built', 'Gr Liv Area', 'Overall Qual', 'Full Bath'] + \
                   [col for col in df_encoded.columns if 'House Style_' in col]
X_exp1 = df_encoded[feature_cols_exp1]
y_exp1 = df_encoded['Garage']

X_train_exp1, X_test_exp1, y_train_exp1, y_test_exp1 = train_test_split(X_exp1, y_exp1, test_size=0.2, random_state=42)

# De boom dieper laten groeien (max_depth=10) en rekenmethode aanpassen (entropy)
dt_model_exp1 = DecisionTreeClassifier(max_depth=10, min_samples_split=2, criterion='entropy', random_state=42)
dt_model_exp1.fit(X_train_exp1, y_train_exp1)

y_pred_exp1 = dt_model_exp1.predict(X_test_exp1)
accuracy_exp1 = accuracy_score(y_test_exp1, y_pred_exp1)
conf_matrix_exp1 = confusion_matrix(y_test_exp1, y_pred_exp1)

print("Resultaat Experiment 1 (Garage):")
print(f"Accuracy: {accuracy_exp1 * 100:.2f}%")
print("Confusion Matrix:\n", conf_matrix_exp1)
```

```python
# Target is nu een cijfer 1 t/m 10 (Multi-class classificatie)
df_encoded2 = pd.get_dummies(df, columns=['Neighborhood'], drop_first=True)
df_encoded2['Total Bsmt SF'] = df_encoded2['Total Bsmt SF'].fillna(0)

y2 = df_encoded2['Overall Qual']
feature_cols2 = ['Gr Liv Area', 'Year Built', 'Total Bsmt SF'] + [col for col in df_encoded2.columns if 'Neighborhood_' in col]
X2 = df_encoded2[feature_cols2]

X_train2, X_test2, y_train2, y_test2 = train_test_split(X2, y2, test_size=0.2, random_state=42)

# Boom redelijk ondiep houden om te kijken naar de algemene patronen
dt_model2 = DecisionTreeClassifier(max_depth=6, min_samples_split=5, random_state=42)
dt_model2.fit(X_train2, y_train2)
y_pred2 = dt_model2.predict(X_test2)

accuracy2 = accuracy_score(y_test2, y_pred2)
conf_matrix2 = confusion_matrix(y_test2, y_pred2)

print("Resultaten Baseline Kwaliteit:")
print(f"Accuracy: {accuracy2 * 100:.2f}%")
print("Confusion Matrix:\n", conf_matrix2)
```

```python
df_encoded_exp2 = pd.get_dummies(df, columns=['Neighborhood', 'House Style'], drop_first=True)
df_encoded_exp2['Total Bsmt SF'] = df_encoded_exp2['Total Bsmt SF'].fillna(0)

feature_cols_exp2 = ['Gr Liv Area', 'Year Built', 'Total Bsmt SF', 'Full Bath', 'Bedroom AbvGr'] + \
                   [col for col in df_encoded_exp2.columns if 'Neighborhood_' in col or 'House Style_' in col]

X_exp2 = df_encoded_exp2[feature_cols_exp2]
y_exp2 = df_encoded_exp2['Overall Qual']

X_train_exp2, X_test_exp2, y_train_exp2, y_test_exp2 = train_test_split(X_exp2, y_exp2, test_size=0.2, random_state=42)

# Boom te diep laten doorgroeien (max_depth=12)
# Hierdoor leert het model de train-data uit zijn hoofd en presteert hij slechter op test-data
dt_model_exp2 = DecisionTreeClassifier(max_depth=12, min_samples_split=4, criterion='entropy', random_state=42)
dt_model_exp2.fit(X_train_exp2, y_train_exp2)
y_pred_exp2 = dt_model_exp2.predict(X_test_exp2)

accuracy_exp2 = accuracy_score(y_test_exp2, y_pred_exp2)
conf_matrix_exp2 = confusion_matrix(y_test_exp2, y_pred_exp2)

print("Resultaat Experiment 2 (Overfitting op kwaliteit):")
print(f"Accuracy: {accuracy_exp2 * 100:.2f}%")
print("Confusion Matrix:\n", conf_matrix_exp2)
```
