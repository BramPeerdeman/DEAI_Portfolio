```python
import sqlite3

conn_sdm = sqlite3.connect('BikeToDrive.db')
cursor = conn_sdm.cursor()

print("--- Start Data Mutaties voor Stap 5 ---")

print("1. SCD1: Woonplaats van monteur 1 aanpassen...")
cursor.execute("""
    UPDATE Monteur
    SET woonplaats = 'Test Woonplaats SCD1'
    WHERE monteurnr = 1;
""")

print("2. SCD2: Adres van klant 1 aanpassen...")
cursor.execute("""
    UPDATE Klant
    SET adres = 'Nieuw Adres SCD2 Testmmm'
    WHERE klantnr = 1;
""")

cursor.execute("SELECT MAX(fiets_verkoopnr) FROM Fiets_Verkoop")
max_fv = cursor.fetchone()[0]
new_fv_id = (max_fv or 0) + 1

cursor.execute("SELECT fietsnr FROM Fiets LIMIT 1")
fiets_id = cursor.fetchone()[0]

print("3. Fact: Nieuwe Fiets_Verkoop toevoegen voor klant 1...")
cursor.execute("""
    INSERT INTO Fiets_Verkoop (fiets_verkoopnr, datum, aantal, verkoopprijs, klant, fiets, monteur)
    VALUES (?, date('now'), 1, 999.99, 1, ?, 1);
""", (new_fv_id, fiets_id))

conn_sdm.commit()
conn_sdm.close()

print("--- Testdata succesvol doorgevoerd in BikeToDrive.db! ---")
```
